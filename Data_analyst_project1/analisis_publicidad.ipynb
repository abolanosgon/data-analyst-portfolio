```python
import pandas as pd
print("✅ pandas funciona correctamente")
```

```python
# Cargar el archivo CSV
df = pd.read_csv("data.csv")

# Ver las primeras filas
df.head()
```

```python
df['reporting_start'] = pd.to_datetime(df['reporting_start'], format='%d/%m/%Y')
df['reporting_end'] = pd.to_datetime(df['reporting_end'], format='%d/%m/%Y')
df.head()
```

```python
# Recuento de valores nulos por columna
df.isnull().sum()
```

```python
#CTR (Click Through Rate)
df['CTR'] = (df['clicks'] / df['impressions']) * 100
df.head()
```

```python
#CPC (Costo por Clic)
df['CPC'] = df.apply(lambda row: row['spent'] / row['clicks'] if row['clicks'] > 0 else 0, axis=1)
df.head()
```

```python
#CPA (Costo por Conversión Aprobada)
df['CPA'] = df.apply(lambda row: row['spent'] / row['approved_conversion'] if row['approved_conversion'] > 0 else 0, axis=1)
df.head()
```

```python
# Tasa de conversion
df['conversion_rate'] = df.apply(lambda row: (row['approved_conversion'] / row['clicks']) * 100 if row['clicks'] > 0 else 0, axis=1)
df.head()
```

```python
df[['CTR', 'CPC', 'CPA', 'conversion_rate']].head()
```

```python
df.to_csv("data_limpio.csv", index=False)
```
